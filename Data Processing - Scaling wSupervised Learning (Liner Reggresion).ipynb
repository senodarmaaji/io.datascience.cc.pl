```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.DataFrame({'WEIGHT': [15, 18, 12, 10],
                  'PRICE': [1, 3, 2, 5]})
```

```python
index = ['Orange', 'Apple', 'Banana', 'Grape']
```

```python
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df1 = pd.DataFrame(scaler.fit_transform(df),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape'])
```

```python
print(df1)
```

```python
import matplotlib.pyplot as plt

# Create the first scatter plot
ax = df.plot.scatter(x='WEIGHT', y='PRICE', color=['red', 'green', 'blue', 'yellow'],
                     marker='*', s=80, label='BEFORE SCALING')

# Create the second scatter plot
df1.plot.scatter(x='WEIGHT', y='PRICE', color=['red', 'green', 'blue', 'yellow'],
                 marker='o', s=60, label='AFTER SCALING', ax=ax)

# Add horizontal and vertical lines
plt.axhline(0, color='red', alpha=0.2)
plt.axvline(0, color='red', alpha=0.2)

# Display the plot
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df[['WEIGHT', 'PRICE']]
y = df['PRICE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  # Adjust the test_size and random_state as desired
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
rmse = np.sqrt(mse)
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Squared Error:", mse)
```

```python
print("Root Mean Squared Error:", rmse)
```

```python
print("R-squared Score:", r2)
```
